```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()

ServiceKey = os.getenv('OPENAPI_SERVICE_KEY')
start_page = 1

# API 기본 설정
url = 'http://apis.data.go.kr/1471000/DrbEasyDrugInfoService/getDrbEasyDrugList'
params = {
    'serviceKey': ServiceKey,
    'pageNo': str(start_page),  # 초기 페이지 번호
    'numOfRows': '100',
    'entpName': '',
    'itemName': '',
    'itemSeq': '',
    'efcyQesitm': '',
    'useMethodQesitm': '',
    'atpnWarnQesitm': '',
    'atpnQesitm': '',
    'intrcQesitm': '',
    'seQesitm': '',
    'depositMethodQesitm': '',
    'openDe': '',
    'updateDe': '',
    'type': 'json'  # JSON 형식으로 요청
}

# DataFrame 초기화
all_data = []

while True:
    # 현재 페이지 번호를 파라미터에 설정
    params['pageNo'] = str(start_page)
    
    # API 호출
    response = requests.get(url, params=params)

    # 응답 내용 확인
    if response.status_code == 200:
        # 응답을 JSON으로 변환
        data = response.json()

        # items 추출
        items = data['body']['items']
        
        # items가 비어 있으면 종료
        if not items:
            break
        
        # 데이터 추가
        all_data.extend(items)
        
        # 페이지 번호 증가
        start_page += 1
    else:
        print(f"Error: {response.status_code}")
        break

# DataFrame으로 변환
df = pd.DataFrame(all_data)

# 결과 출력
df.info()
```
