```python
import pandas as pd

consumed = pd. read_csv('./data/4.국민건강보험공단_의약품처방정보_(별첨)주성분코드변환내역.csv', encoding='cp949')
consumed.info()
```

```python
consumed.rename(columns={'주성분코드':'일반명코드(성분명코드)'}, inplace=True)
```

```python
mapping = pd.read_json('./data/e약은요_약성포함.json')
mapping.info()
```

```python
merged_df = pd.merge(consumed, mapping, on='일반명코드(성분명코드)', how='left')

# 결과 출력
merged_df[['일반명코드(성분명코드)', '제품명', '일반명']].info()
```
