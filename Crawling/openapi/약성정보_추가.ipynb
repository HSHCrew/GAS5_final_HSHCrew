```python
import pandas as pd

consumed = pd. read_json('./data/e약은요_일반명코드.json')
consumed.head(1)
```

```python
mapping = pd.read_csv('./data/건강보험심사평가원_약가마스터_의약품주성분_20231001.csv', encoding='cp949')
mapping.head(2)
```

```python
mapping['일반명코드(성분명코드)'] = mapping.일반명코드
```

```python
merged_df = pd.merge(consumed, mapping, on='일반명코드(성분명코드)', how='left')

# 결과 출력
print(merged_df[['일반명코드(성분명코드)','일반명', '투여', '제품명']])
```

```python
merged_df[merged_df.제품명!=None].info()
```

```python
merged_df.to_json('./data/e약은요_약성포함.json')
```
